# Prática 1: Modulação AM-DSB
## 1.1 - Expressandos os três casos clássicos da modulação: sobremodulação; submodulação e 100% modulado

```python
# Importa bibliotecas
import numpy as np
import matplotlib.pyplot as plt
from scipy import fftpack

# Parâmetros do sinal
Ac = 2                                                         # Amplitude da portadora
Mu = 1                                                       # Índice de modulação
fc = 25000                                                     # Frequência da portadora Hz
fm = 2000
N = 1000
Ts = 1e-6                                                      # Tempo de amostragem pequeno (modelar sinal contínuo)
t = np.arange(N)*Ts
s = Ac*(1+Mu*np.cos(2*np.pi*fm*t))*np.cos(2*np.pi*fc*t)
Mu=1.5
s1=Ac*(1+Mu*np.cos(2*np.pi*fm*t))*np.cos(2*np.pi*fc*t)
Mu=0.4
s2=Ac*(1+Mu*np.cos(2*np.pi*fm*t))*np.cos(2*np.pi*fc*t)
```

 __No cóodigo acima, *s* representa uma modulação 100%, *s1* representa uma sobremodulação, com $\mu = 1,5$ e *s2* representa uma submodulação, com $\mu=0,4$__

__Para visualizar o que ocorre com esses três tipos de modulação, plotamos cada uma delas:__

```python
# Gráfico do AM-DSB 100% modulado
plt.figure(1,[10,7])
plt.subplot(311)
plt.plot(t,s)
plt.title("Sinal AM 100% modulaçao")
plt.xlabel("Tempo [s]")
plt.ylabel("Amplitude")

# Gráfico do AM-DSB sobremodulado
plt.subplot(312)
plt.title("Sinal AM Sobremodulado")
plt.xlabel("Tempo[s]")
plt.ylabel("Amplitude")
plt.plot(t,s1)
#Grafico do AM-DSB submodulado
plt.subplot(313)
plt.title("Sinal AM Submodulado")
plt.xlabel("Tempo[s]")
plt.ylabel("Amplitude")
plt.plot(t,s2)
plt.tight_layout()
plt.show()
```

Em uma modulação cujo índice é menor do que 100% não faz com que o sinal varie em toda sua amplitude. Pode ocorrer ainda que a modulação supere os 100%, caso em que teremos instantes em que o sinal se anula, e ocorre inversão de fase.
Um sinal não deve ser modulado em mais do que 100%, pois além de ocorrerem distorções dos sinais, são geradas bandas laterais fora das faixas designadas causando interferências.[1] 



## 1.2 - Calcular a potência do sinal AM-DSB

Considerando um sinal genérico $x(t) = A\cos(2\pi ft +\phi)$, temos que a potência desse sinal é expressa da forma $P_x = \frac{A^2}{2}$. Por sua vez, um sinal AM tem como expressão matemática:  <br>
$s_{AM-DSB}(t)=A_{c}\cdot cos(2\pi f_{c}t)+ \frac{A_c \cdot m_a}{2} \cdot cos[2\pi (f_{c}-f_m)t]+ \frac{A_c \cdot m_a}{2} \cdot cos[2\pi (f_{c}+f_m)t]$ <br>
Dessa forma, a potência de um sinal AM é determinado levando em conta as amplitudes de cada um dos termos senoidais dessa expressão: $P_s = \frac{A_c^2}{2} + \frac{A_c^2m_a^2}{8} + \frac{A_c^2m_a^2}{8}$, que irá servir como valor teórico de referência aos outros métodos de cálculo de potência.

__Criando o sinal AM e calculando seu valor teórico de potência:__

```python
#Importando bibliotecas
import numpy as np
import matplotlib.pyplot as plt
from scipy import fftpack
import scipy.signal as sci


# Parâmetros do sinal
Ac = 2                                                      # Amplitude da portadora
Mu = 1                                                      # Índice de modulação
fc = 25000                                                  # Frequência da portadora Hz
fm = 2000
N = 1000
Ts = 1e-6                                                   # Tempo de amostragem pequeno (modelar sinal contínuo)
Fs=1/Ts
t = np.arange(N)*Ts
s = Ac*(1+Mu*np.cos(2*np.pi*fm*t))*np.cos(2*np.pi*fc*t)
#Calculando o valor teorico:
P_teorico=Ac**2/2+ (Ac**2*Mu**2)/8+ (Ac**2*Mu**2)/8
print('Potencia teorica = ',P_teorico)
```

__A seguir, foram realizados dois métodos de cálculo de potência, pelo valor das amostras e pelo valor da autocorrelação:__
* $P = \frac{1}{N}\sum_{n=0}^{N-1} \left |  x[n]  \right |^{2}$
* $P = R_{xx}(0)$, em que $R_{xx}$ é a função de autocorrelação

```python
#Calculando a potencia pelas amostras
Px_tempo =(np.linalg.norm(s)**2)/N   # Cálculo da potência no tempo
print('Potência via amostras no tempo = ',Px_tempo)
#Conferindo via correlaçao
Rxx=np.correlate(s,s,'full')/N  # Estima a autocorrelaçao de x(n)
Px_Rxx = Rxx[N-1]                    # Cálculo da potência duas bandas via autocorrelação
print('Potência via autocorrelação = ',Px_Rxx)
```

__Como se pode ver, os três métodos de determinação do valor de potência geram o mesmo valor__<br><br>
Além disso, podemos ver o periodograma de uma modulação AM-DSB:

```python
from spectrum.window import Window
hamming = Window(len(s),name='hamming')
f,pxx = sci.periodogram(s,window=hamming.data,fs=Fs,nfft=len(s),scaling='spectrum')
#pwrest = pxx.max()
#idx = pxx.argmax()
plt.plot(f/1e3,pxx)
plt.title('Periodograma')
plt.xlabel('Frequencia [kHz]')
plt.ylabel('Potencia [W]')
plt.tight_layout()
plt.grid()
plt.axis([20,30,0,Ac**2])
plt.show()
```

Diferente do periodograma trabalhado nas práticas anteriores, este não apresenta uma raia única onde toda a potência se concntra. Pode-se ver que a potência está distribuida na faixa em torno da portadora - 25 kHz - com a largura do sinal modulante - 2 kHz.

# Prática 2: Recepção AM-DSB (detecção de envoltória)

__Nessa prática, foi analisada a qualidade do detector de envoltórias a partir de variadas constantes de tempo do circuito capacitivo: $\tau = 10^{-4}$, $\tau = 2\times 10^{-4}$ e $\tau = 10^{-3}$__

```python

import numpy as np
from matplotlib import pyplot as plt
from scipy import fftpack
def calc_envoltoria(Ns,tau,x_AM): #funcao para determinar o valor da envoltoria (parametros: numero de pontos,valor da constante de tempo e sinal de entrada (já retificado))
    fx_env=np.zeros(len(x_AM))
    out=-1
    for i in range(Ns):
        inp = x_AM[i]
        if inp>=out:
            out = inp            # Caso 1: x_am(t) > Vc(t) (carga do capacitor)
        else:
            out *= (1-Ts/tau)    # Caso 2: x_am(t) < Vc(t) (descarga do capacitor)
        fx_env[i] = out
    return fx_env

    
tau = 1e-4                                                      # Constante de tempo do detector de envelope
tau1= 2e-4
tau2=4e-3
Ts=1e-6                                                         # Definição do período
t = np.arange(1000)*Ts                                          # Definição do vetor tempo
fc = 10000                                                      # Frequência da portadora.
fm = 1500                                                       # Frequência do sinal
Mu = 0.6                                                        # Índice de modulaçao.
Ac = 1.0
x_AMo = Ac*(1.0+Mu*np.cos(2*np.pi*fm*t))*np.cos(2*np.pi*fc*t);  # Onda Modulada AM-DSB

x_envIdeal = np.abs(Ac*(1.0+Mu*np.cos(2*np.pi*fm*t)))           # Envoltória ideal

# Detector de envoltória
x_AM = x_AMo*(x_AMo>0)                                          # Efeito do diodo (semiciclo positivo)
x_env = np.zeros(len(x_AM))
Ns = len(x_AM)
x_env=calc_envoltoria(Ns,tau,x_AM)                             #X_AM já retificada 
x_env1=calc_envoltoria(Ns,tau1,x_AM)
x_env2=calc_envoltoria(Ns,tau2,x_AM)
#Calculo do erro medio quadrático:
mse = ((x_envIdeal - x_env)**2).mean(axis=None)
ms1 = ((x_envIdeal - x_env1)**2).mean(axis=None)
ms2 = ((x_envIdeal - x_env2)**2).mean(axis=None)
print("mse para tau = 1e-4: ", mse)
print("mse para tau = 2e-4: ", ms1)
print("mse para tau = 1e-3: ", ms2)
```

__Numericamente, vemos que o circuito que melhor regenera a onda modulante é o de constante de tempo $\tau = 2 \times 10^{-4}$, pois apresenta  o menor MSE. Vamos visualizar isso:__

```python
plt.figure(1,[10,7])
#plt.subplot(313)
plt.plot(t*1000,x_envIdeal)
plt.plot(t*1000,x_env)  
plt.plot(t*1000,x_env1)
plt.plot(t*1e3,x_env2)                           
plt.title("Envoltória recuperada")
plt.ylabel("Amplitude")
plt.xlabel("Tempo(ms)")
plt.legend(["envoltoria","Tau","Tau1","Tau2"])

plt.show()
```

__O capacitor que determina a  constante $10^{-4}$ descarrega muito rápido,  e nesses momentos se distancia bastante da envoltória ideal. Já o capacitor que determina a constante $10^{-3}$ acumula muita carga e demora para descarregar, não acompanhando a senoide da envoltória. O capacitor de constante $2\times10^{-4}$ oferece uma resposta intermediária: não descarrega tão rápido quanto o primeiro, e por isso não fica tão longe da envoltória, nem é tão lento quanto o último, o que geraria uma resposta grosseira__

# Prática 8 - Demodulação AM-DSB com Transformada de Hilbert

Determinaçao dos parâmetros da modulação:

```python
import numpy as np
from scipy.signal import hilbert
from scipy import fftpack 
import matplotlib.pyplot as plt

# Parâmetros
fs = 600                                                  # frequencia de amostragem
Ts = 1/fs
t = np.arange(600)*Ts
fm = 3                                                   #Frequência da onda modulante
fc = 50                                                  # Frequência da portadora
Mu=0.7                                                   #indice de modulacao
Ac=1                                                    #amplitude da portadora (arbitrário)
x_AM = Ac*(1.0+Mu*np.cos(2*np.pi*fm*t))*np.cos(2*np.pi*fc*t);  # Onda Modulada AM-DSB
carrier_org=Ac*np.cos(2*np.pi*fc*t)                     #formula da portadora para posterior comparação
m=Mu*Ac*np.cos(2*np.pi*fm*t)                            # formula da modulante para posterior comparação  
```

Aplicando a transformada de Hilbert:

```python
# Hilbert
z= hilbert(x_AM)                                             # Sinal analítico (real + imaginário) da onda modulada
x_env = np.abs(z)                                            # modulo para recuperação da envoltoria
inst_phase = np.unwrap(np.angle(z))                          #inst phase para recuperação da portadora
carrier_reg=np.cos(inst_phase)                               #portadora regenerada
```

Como mostrado no HandsOn, para recuperar a envoltória da AM, trabalhamos com o módulo da transformada de Hilbert. Na modulação AM, a frequência da onda modulada é a mesma da portadora - onda modulante só atua na amplitude. Dessa forma, a fase, ou frequência da transformada de hilbert é a própria frequência da portadora.

```python
#Analise do espectro
lfft = len(x_AM)                                           # Comprimento do sinal DSB
lfft = int(2**np.ceil(np.log2(lfft)))                      # Comprimento do sinal DSB potência de dois
freq=np.arange(-fs/2,fs/2,fs/lfft)                         # Eixo das frequências

X_m=np.abs(fftpack.fftshift(fftpack.fft(m,lfft)))/lfft     # FFT da onda modulante original
X_AM = np.abs(fftpack.fftshift(fftpack.fft(x_AM,lfft)))/lfft # FFT do sinal modulado
X_demodulado = np.abs(fftpack.fftshift(fftpack.fft(x_env,lfft)))/lfft # FFT do sinal demodulado pela transformada de Hilbert, ou: FFT da envoltória
```

Gerando gráficos:
- Sinal demodulado pela Transformada de Hilbert

```python
# Gráficos
plt.figure(1)
plt.title('Demodulação AM com transformada de Hilbert')
plt.xlabel('Tempo [s]')
plt.ylabel('Amplitude')
plt.plot(t,x_AM,t,x_env)
plt.ylim([-2,2])
plt.xlim([0,1])
plt.legend(['s(t)','envoltoria'])
plt.show()
```

 - Espectro do Sinal Modulante e do Sinal Demodulado:

```python
plt.figure(2)
plt.title('Espectro Demodulação')
plt.xlabel('Frequência [Hz]')
plt.ylabel('Amplitude [abs]')
plt.plot(freq,X_m,freq,X_demodulado)
plt.xlim([-10,10])
plt.legend(['M(f)','M(f)_reg'])
plt.show()
```

Observamos que o sinal regenerado possui uma componente DC gerado pelo processo de detecção de envoltória, que poderia ser facilmente removido na prática, ou com um circuito de desacoplamento, ou um subtrator. O ponto principal a ser visto é a coincidência das compomentes na faixa de frequência do sinal modulante - 3Hz. Indicando que o processo de demodulação foi bem sucedido.

 - Escpectro do sinal modulado:

```python
plt.figure(3)
plt.title('Espectro Modulação')
plt.xlabel('Frequência [Hz]')
plt.ylabel('Amplitude [abs]')
plt.plot(freq,X_AM)
plt.xlim([-80,80])
plt.legend('S(f)')
plt.show()
```

Observa-se a raia de maior amplitude na faixa de 50Hz (portadora) e as raias laterais $f_c + f_m$ e $f_c - f_m$ típicos do AM-DSB

- Regeneração da portadora:

```python
plt.figure(4)
plt.title('Portadora')
plt.xlabel('Frequência [Hz]')
plt.ylabel('Amplitude')
plt.plot(t,carrier_org,t,carrier_reg)
plt.legend(['c(t)','c(t)_reg'])
plt.show()
```

Com a coincidência entre a portadora real e a portadora regenerada por meio da fase do sinal analítico, observa-se a correta execução da transformada de Hilbert no experimento
